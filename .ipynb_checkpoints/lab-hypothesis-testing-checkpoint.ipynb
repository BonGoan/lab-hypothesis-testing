In this challenge, we will be working with pokemon data. The data can be found here:

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

```python
df['Type 1'].unique()
```

```python
df['Type 2'].unique()
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#code here
#hypothesis
#H0: mu Dragon !> mu Grass
#H1: mu Dragon > mu Grass
#with alpha = 0.05

dragon_hp = pd.concat([df[df['Type 1'] == 'Dragon']['HP'], df[df['Type 2'] == 'Dragon']['HP']])
grass_hp = pd.concat([df[df['Type 1'] == 'Grass']['HP'],df[df['Type 2'] == 'Grass']['HP']])

st.ttest_ind(dragon_hp, grass_hp, equal_var=False, alternative='greater')




```

```python
#Interpretation
#t-test = 4.097
#p-value = 5.0908e-05
#degress of freedom = 77.581

#since the p value is very small, we reject the null hypothesis
#This means that Dragon-type Pokémon have, on average, significantly higher HP than Grass-type Pokémon.
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
#code here
#hypothesis
#H0: mu Legendary = mu Non Legendary
#H1: mu Legendary != > mu Non Legendary
#with alpha = 0.05

legendary = df[df['Legendary'] == True]
non_legendary = df[df['Legendary'] == False]

stats_to_test = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
t_test_results = {}

for stat in stats_to_test:
    t_stat, p_value = st.ttest_ind(legendary[stat], non_legendary[stat], equal_var=False)
    t_test_results[stat] = (t_stat, p_value)

for stat, (t_stat, p_value) in t_test_results.items():
    print(f"{stat}: t-statistic = {t_stat:.2f}, p-value = {p_value:.5f}")

if p_value < 0.05:
        print(f"  -> Significant difference in {stat} between Legendary and Non-Legendary Pokémon")
else:
    print(f"  -> No significant difference in {stat} between Legendary and Non-Legendary Pokémon")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
school_coordinates = (-118, 37)
hospital_coordinates = (-122, 34)
```

```python
#euclidean distance formula:
#Calculate the difference between the two longitudes: lon1 - lon2.
#Square the result: (lon1 - lon2)**2.
#Do the same for latitudes: (lat1 - lat2)**2.
#Add the squares of the differences.
#Take the square root of the sum using np.sqrt().

def euclidean_distance(lon1, lat1, lon2, lat2):
    return np.sqrt((lon1 - lon2)**2 + (lat1 - lat2)**2)

df['distance_to_school'] = euclidean_distance(df['longitude'], df['latitude'], school_coordinates[0], school_coordinates[1])
df['distance_to_hospital'] = euclidean_distance(df['longitude'], df['latitude'], hospital_coordinates[0], hospital_coordinates[1])

df['close_to_school_or_hospital'] = (df['distance_to_school'] < 0.50) | (df['distance_to_hospital'] < 0.50)
df
```

```python
#euclidean distance formula:
#Calculate the difference between the two longitudes: lon1 - lon2.
#Square the result: (lon1 - lon2)**2.
#Do the same for latitudes: (lat1 - lat2)**2.
#Add the squares of the differences.
#Take the square root of the sum using np.sqrt().

def euclidean_distance(lon1, lat1, lon2, lat2):
    return np.sqrt((lon1 - lon2)**2 + (lat1 - lat2)**2)

df['distance_to_school'] = euclidean_distance(df['longitude'], df['latitude'], school_coordinates[0], school_coordinates[1])
df['distance_to_hospital'] = euclidean_distance(df['longitude'], df['latitude'], hospital_coordinates[0], hospital_coordinates[1])

df['close_to_school_or_hospital'] = (df['distance_to_school'] < 0.50) | (df['distance_to_hospital'] < 0.50)

close_group = df[df['close_to_school_or_hospital'] == True]
far_group = df[df['close_to_school_or_hospital'] == False]

#H0: no difference in the median prices of the close and far group
#H1: there is a differnece between the median prices of the close and far group
#alpa = 0.05

t_stat, p_value = st.ttest_ind(close_group['median_house_value'], far_group['median_house_value'], equal_var=False)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.5f}")
```

#Interpretation
P-value is very very small. so we reject the null hypothesis and accept the alternate hypothesis.
This mean there is significant difference between the median prices of the close and far groups.
T-statistic negative value indicates that the houses close to a hospital have significantly lower median house prices that those far away.
Hence, we can conclude:
Houses close to a school or hospital appear to be less expensive.
